```python
 data.drop(['Row ID', 'Customer ID', 'Postal Code'], axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.groupby('Category')['Sales'].sum().sort_values(ascending=True)
```

```python
region_sales = data.groupby('Region')['Sales'].sum().sort_values()
region_sales.plot(kind='bar', title='Sales by Region', color='purple')
plt.ylabel('Total Sales')
plt.show()

region_sales = data.groupby('Region')['Sales'].sum().reset_index().sort_values('Sales')

sns.barplot(data=region_sales, x='Region', y='Sales', color='purple')
plt.title('Sales by Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=data, x='Discount', y='Profit', hue='Category')
plt.title("Profit vs Discount by Category")
plt.show()
```

```python
subcat_profit = data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(10)
subcat_profit.plot(kind='barh', title='Top 10 Profitable Sub-Categories', color='green')
plt.xlabel('Profit')
plt.show()

subcat_profit = data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).reset_index()

sns.barplot(data=subcat_profit, y='Sub-Category', x='Profit', color='green')
plt.title('Top 10 Profitable Sub-Categories')
plt.xlabel('Profit')
plt.ylabel('Sub-Category')
plt.tight_layout()
plt.show()
```
